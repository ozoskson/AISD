10) –ù–∞–ø–∏—Å–∞—Ç—å 4 —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ä–æ–∂–¥–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ (–ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –∏ —É–±—ã–≤–∞–Ω–∏—é), —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∏ –ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é. –ö–∞–∂–¥–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞.


```python
import random
def generate_sorted_list(leng): #–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤
    return [i for i in range(1,leng+1)]
print(generate_sorted_list(10))
def generate_unsorted_list(leng): #–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è
    return [i for i in range(leng, 0, -1)]
print(generate_unsorted_list(10))
def generate_random_list(leng): #–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª
    return [random.randint(1, 10) for _ in range(leng)]
print(generate_random_list(10))
def generate_non_sorted_list(leng): #–ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é
    arr = list(range(1, leng+1))
    tmp = arr[-2]
    arr[-2] = arr[-1]
    arr[-1] = tmp
    return arr
print(generate_non_sorted_list(10))
```

–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏. –ü—Ä–æ—Ü–µ–¥—É—Ä–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –∏ –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ—Ä–æ–∂–¥–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ f_gen(list_len), –≥–µ–Ω–µ—Ä–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫ –∑–∞–¥–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–æ–µ–Ω–∏–π —Ç–µ—Å—Ç–∞ –∏ —Å–ø–∏—Å–æ–∫ –¥–ª–∏–Ω –∏—Å–ø–æ–ª—å–∑—É—é—â–∏—Ö—Å—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è. –î–ª—è –∫–∞–∂–¥–æ–π –¥–ª–∏–Ω—ã —Å–ø–∏—Å–∫–∞ —Ñ—É–Ω–∫—Ü–∏—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫.


```python
import statistics

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
def test_sorting_algorithm(sort_func, list_generator, repetitions, lengths):
    results = {}  # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ª–æ–≤–∞—Ä—å –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    for length in lengths:  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º –¥–ª–∏–Ω–∞–º —Å–ø–∏—Å–∫–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
        comparisons = []  # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
        swaps = []  # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
        for i in range(repetitions):  # –ü–æ–≤—Ç–æ—Ä—è–µ–º —Ç–µ—Å—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑
            # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ —Å –ø–æ–º–æ—â—å—é –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ list_generator
            lst = list_generator(length)
            # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –¥–ª—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞
            comparisons_count, swaps_count = sort_func(lst) 
            comparisons.append(comparisons_count)  # –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –≤ —Å–ø–∏—Å–æ–∫
            swaps.append(swaps_count)  # –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –≤ —Å–ø–∏—Å–æ–∫
        # –ó–∞–ø–∏—Å—ã–≤–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–µ—Å—Ç–∞ –≤ —Å–ª–æ–≤–∞—Ä—å
        results[f'–î–ª—è –¥–ª–∏–Ω—ã —Å–ø–∏—Å–∫–∞ {length}'] = {
            '–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π': min(comparisons),  
            '–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π': statistics.mean(comparisons), 
            '–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π': max(comparisons), 
            '–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫': min(swaps),  
            '–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫': statistics.mean(swaps),  
            '–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫': max(swaps) 
        }
    return results  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ —Ç–µ—Å—Ç–∞

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø—É–∑—ã—Ä—å–∫–æ–º –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ä–∞–±–æ—Ç—ã test_sorting_algorithm
def bubble_sort(lst):
    comparisons = 0  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
    swaps = 0  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
    n = len(lst)  # –î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞
    for i in range(n):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ø–∏—Å–∫–∞
        for j in range(n - i - 1):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º –¥–æ n-i-1
            comparisons += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
            if lst[j] > lst[j + 1]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –±–æ–ª—å—à–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ
                lst[j], lst[j + 1] = lst[j + 1], lst[j]  # –ú–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏
                swaps += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
    return comparisons, swaps  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫

# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ "–ø—É–∑—ã—Ä–µ–∫" 
print("–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ø—É–∑—ã—Ä–µ–∫:")
print("=" * 50) # –ü—Ä–æ—Å—Ç–æ –¥–ª—è –∫—Ä–∞—Å–∏–≤–æ–≥–æ –≤—ã–≤–æ–¥–∞
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞:")
list_lengths = [10, 100, 1000]  # –ø—Ä–∏–º–µ—Ä —Å–ø–∏—Å–∫–∞ –¥–ª–∏–Ω –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
repetitions = 10  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π —Ç–µ—Å—Ç–∞
results = test_sorting_algorithm(bubble_sort, generate_sorted_list, repetitions, list_lengths)
print(results)
```

12. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø—É–∑—ã—Ä—å–∫–æ–º: "–∫–∞–º–µ—à–µ–∫-–ø—É–∑—ã—Ä–µ–∫" - —Å —á–µ—Ä–µ–¥—É—é—â–∏–º–∏—Å—è –ø—Ä–æ—Ö–æ–¥–∞–º–∏, –ø—Ä–∏–≤–æ–¥—è—â–∏–º–∏ –∫ "–≤—Å–ø–ª—ã—Ç–∏—é" —Å–∞–º–æ–≥–æ –±–æ–ª—å—à–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –∏ "–ø–æ–≥—Ä—É–∂–µ–Ω–∏—é" —Å–∞–º–æ–≥–æ –º–∞–ª–µ–Ω—å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞.

```python
def stones_bubble_sort(arr):
    while True:
        flag = 0
        comparisons = 0
        swaps = 0
        for index in range(len(arr) - 1):
            index_from_last = len(arr) - index - 1 
            comparisons += 1
            if arr[index] > arr[index+1]:
                arr[index],arr[index+1] = arr[index+1],arr[index] 
                flag = 1
                swaps += 1
            if arr[index_from_last] < arr[index_from_last - 1]:
                arr[index_from_last], arr[index_from_last - 1] = arr[index_from_last - 1], arr[index_from_last]
                flag = 1
                swaps += 1
        if flag == 0:
            return comparisons, swaps
```


15) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—ã–±–æ—Ä–∞ –∏–∑ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤ (–æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ –≤ –ª–µ–∫—Ü–∏–∏). –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º.


```python
def modified_selection_sort(arr):
    sublist_size = 3  # –†–∞–∑–º–µ—Ä –ø–æ–¥—Å–ø–∏—Å–∫–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 3, –∫–∞–∫ –∑–¥–µ—Å—å, –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–º)
    comparisons = 0  # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
    swaps = 0  # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
    n = len(arr)  # –î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞

    for i in range(0, n, sublist_size):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º —Ä–∞–≤–Ω—ã–º —Ä–∞–∑–º–µ—Ä—É –ø–æ–¥—Å–ø–∏—Å–∫–∞
        end = min(i + sublist_size, n)  # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–Ω—Ü–∞ –ø–æ–¥—Å–ø–∏—Å–∫–∞. –≠—Ç–æ —á—Ç–æ–±—ã –Ω–µ –≤—ã–π—Ç–∏ –∑–∞ –∫—Ä–∞—è —Å–ø–∏—Å–∫–∞ 
        for j in range(i, end):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º –ø–æ–¥—Å–ø–∏—Å–∫–∞ –¥–æ –ø—Ä–µ–¥–ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ
            min_index = j  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞
            for k in range(j + 1, end - 1):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º –ø–æ–¥—Å–ø–∏—Å–∫–∞ –æ—Ç —Å–ª–µ–¥—É—é—â–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –¥–æ –∫–æ–Ω—Ü–∞ 
                comparisons += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
                if arr[k] < arr[min_index]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –º–µ–Ω—å—à–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ
                    min_index = k  # –û–±–Ω–æ–≤–ª—è–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏–Ω–¥–µ–∫—Å
            if min_index != j:  # –ï—Å–ª–∏ –Ω–∞–π–¥–µ–Ω–Ω—ã–π –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏
                arr[j], arr[min_index] = arr[min_index], arr[j]  # –ú–µ–Ω—è–µ–º —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç —Å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º
                swaps += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫

    return comparisons, swaps


# –≠—Ç–æ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º –¥–µ–ª–∏—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞ –ø–æ–¥—Å–ø–∏—Å–∫–∏ –∏ –Ω–∞—Ö–æ–¥–∏—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –∫–∞–∂–¥–æ–º –ø–æ–¥—Å–ø–∏—Å–∫–µ. –ó–∞—Ç–µ–º –æ–Ω –æ–±–º–µ–Ω–∏–≤–∞–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å –ø–µ—Ä–≤—ã–º —ç–ª–µ–º–µ–Ω—Ç–æ–º –ø–æ–¥—Å–ø–∏—Å–∫–∞, –∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å –ø–æ—Å–ª–µ–¥–Ω–∏–º —ç–ª–µ–º–µ–Ω—Ç–æ–º –ø–æ–¥—Å–ø–∏—Å–∫–∞.
print("–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—ã–±–æ—Ä–∞ –∏–∑ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤ :")
print("=" * 50)
results = test_sorting_algorithm(modified_selection_sort, generate_random_list, repetitions, list_lengths)
print(results)
```

16) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –±—ã—Å—Ç—Ä—É—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É —Å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é –ø–æ–¥—Å—á–µ—Ç–∞ –æ–ø–µ—Ä–∞—Ü–∏–π. –°—Ä–∞–≤–Ω–∏—Ç—å –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö (–∏ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö) –¥–∞–Ω–Ω—ã—Ö.

```python
def quick_sort(arr):
    comparisons = 0
    swaps = 0

    stack = [(0, len(arr) - 1)]

    while stack:
        low, high = stack.pop()
        pivot_index = random.randint(low, high)
        pivot = arr[pivot_index]
        arr[pivot_index], arr[high] = arr[high], arr[pivot_index]  # –ø–µ—Ä–µ–º–µ—â–∞–µ–º –æ–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –∫–æ–Ω–µ—Ü
        i = low - 1
        for j in range(low, high):
            comparisons += 1
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                swaps += 1
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        swaps += 1

        # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤ —Å—Ç–µ–∫
        if low < i:
            stack.append((low, i))
        if i + 2 < high:
            stack.append((i + 2, high))

    return comparisons, swaps

import time 

def measure_sorting_time(sort_algorithm, arr):
    start_time = time.time()
    sort_algorithm(arr) 
    end_time = time.time()  
    return end_time - start_time 

print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_random_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())  
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫") 
print("=" * 50)

print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_sorted_list(length)  
    time_taken = measure_sorting_time(quick_sort, lst.copy()) 
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫")
print("=" * 50)


print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_unsorted_list(length)  
    time_taken = measure_sorting_time(quick_sort, lst.copy())  
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫") 
print("=" * 50)
```

17) –£–ª—É—á—à–∏—Ç—å —Ä–∞–±–æ—Ç—É –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –∏ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –∑–∞ —Å—á–µ—Ç –∏–∑–º–µ–Ω–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤—ã–±–æ—Ä–∞ —ç–ª–µ–º–µ–Ω—Ç–∞ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –º–∞—Å—Å–∏–≤–∞.

```python
def quick_sort(arr):
    comparisons = 0
    swaps = 0

    stack = [(0, len(arr) - 1)]

    while stack:
        low, high = stack.pop()
        pivot_index = random.randint(low, high)
        pivot = arr[pivot_index]
        arr[pivot_index], arr[high] = arr[high], arr[pivot_index]  # –ø–µ—Ä–µ–º–µ—â–∞–µ–º –æ–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –∫–æ–Ω–µ—Ü
        i = low - 1
        for j in range(low, high):
            comparisons += 1
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                swaps += 1
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        swaps += 1

        # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤ —Å—Ç–µ–∫
        if low < i:
            stack.append((low, i))
        if i + 2 < high:
            stack.append((i + 2, high))

    return comparisons, swaps
#–ú–µ–Ω—è–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–ª–µ–º–µ–Ω—Ç –Ω–∞ —Ä–∞–Ω–¥–æ–º–Ω—ã–π. –≠—Ç–æ –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å –∏–∑–±–µ–∂–∞—Ç—å —Ö—É–¥—à–µ–≥–æ —Å–ª—É—á–∞—è –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –∏–ª–∏ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, —Ç–∞–∫ –∫–∞–∫ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–±–æ—Ä–∞ "–ø–ª–æ—Ö–æ–≥–æ" –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ —Å–Ω–∏–∂–∞–µ—Ç—Å—è.
# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_random_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫")
print("=" * 50)

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_sorted_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫")
print("=" * 50)

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("=" * 50)
for length in list_lengths:
    lst = generate_unsorted_list(length)
    time_taken = measure_sorting_time(quick_sort, lst.copy())
    print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {length}, –í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {time_taken:.6f} —Å–µ–∫")
print("=" * 50)
```

18) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ–ª—É—á–µ–Ω–∏—è –∏–∑ –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–ø–∏—Å–∫–æ–≤ –æ–¥–Ω–æ–≥–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞.

```python
def merge_sorted_lists(lists):
    result = []  # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    
    while True:  # –ó–∞–ø—É—Å–∫–∞–µ–º –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª 
        min_index = None  # min_index —Ö—Ä–∞–Ω–∏—Ç –∏–Ω–¥–µ–∫—Å —Å–ø–∏—Å–∫–∞ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –ø–µ—Ä–≤—ã–º —ç–ª–µ–º–µ–Ω—Ç–æ–º
        
        # –ò—â–µ–º —Ü–∏–∫–ª–æ–º —Å–ø–∏—Å–æ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π:
        for i, lst in enumerate(lists):  # –ò—Ç–µ—Ä–∏—Ä—É–µ–º—Å—è –ø–æ —Å–ø–∏—Å–∫–∞–º –∏ –∏—Ö –∏–Ω–¥–µ–∫—Å–∞–º. –í i - –∏–Ω–¥–µ–∫—Å—ã —Å–ø–∏—Å–∫–æ–≤, –≤ lst - —Å–ø–∏—Å–∫–∏
            if lst:  # –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –Ω–µ –ø—É—Å—Ç
                if min_index == None or lst[0] < lists[min_index][0]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —Å–ø–∏—Å–æ–∫ –∏–º–µ–µ—Ç –º–µ–Ω—å—à–∏–π –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç
                    min_index = i  # –û–±–Ω–æ–≤–ª—è–µ–º –∏–Ω–¥–µ–∫—Å —Å–ø–∏—Å–∫–∞, —Ö—Ä–∞–Ω—è—â–µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
                    
        if min_index == None:  # –ï—Å–ª–∏ –Ω–µ –Ω–∞–π–¥–µ–Ω –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç
            break  # –ü—Ä–µ—Ä—ã–≤–∞–µ–º —Ü–∏–∫–ª. –≠—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç, –µ—Å–ª–∏ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤–æ –≤—Å–µ—Ö —Å–ø–∏—Å–∫–∞—Ö "–∑–∞–∫–æ–Ω—á–∞—Ç—Å—è"
            
        result.append(lists[min_index][0])  # –î–æ–±–∞–≤–ª—è–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
        lists[min_index] = lists[min_index][1:]  # –£–¥–∞–ª—è–µ–º —ç—Ç–æ—Ç —ç–ª–µ–º–µ–Ω—Ç –∏–∑ —Å–ø–∏—Å–∫–∞, –æ–±–Ω–æ–≤–ª—è—è —Å–ø–∏—Å–æ–∫. –û–Ω –∏–º–µ–µ—Ç 0-–æ–π –∏–Ω–¥–µ–∫—Å, —Ç–∞–∫ –∫–∞–∫
        # —Å–ø–∏—Å–∫–∏ –ø–æ —É—Å–ª–æ–≤–∏—é –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ
    return result  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫


# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
sorted_lists = [[1, 3, 5, 7],
                [2, 4, 6, 8],
                [0, 9, 10]]

sorted_result = merge_sorted_lists(sorted_lists)  # –û–±—ä–µ–¥–∏–Ω—è–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å–ø–∏—Å–∫–∏
print(sorted_result)  # –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
```

19. –ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª–∏–Ω —à–∞–≥–æ–≤ –≤ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –®–µ–ª–ª–∞. 19) –ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª–∏–Ω —à–∞–≥–æ–≤ –≤ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –®–µ–ª–ª–∞. –ù–∞ 2[ùëõ/2^(ùëò+1)]+1
—Ç.–µ. –ø—Ä–∏ –ø–µ—Ä–≤–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –±—É–¥—É—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —à–∞–≥–∏: 2[n/4]+1, 2[n/8]+1, ... , 3, 1 .

```python

def shell_sort_modified(lst):
    n = len(lst)  # –î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
    k = 1  
    b = 0  # b - —Ñ–ª–∞–≥ —Ç–æ–≥–æ, —á—Ç–æ gap –¥–æ—Å—Ç–∏–≥–ª–æ –µ–¥–∏–Ω–∏—Ü—ã 
    gap = 2 * (n // (2 ** (k + 1))) + 1  # gap - —à–∞–≥ –≤ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –®–µ–ª–ª–∞. –§–æ—Ä–º—É–ª–∞ –∏–∑ —É—Å–ª–æ–≤–∏—è
    while gap > 0:  # –ü–æ–∫–∞ –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        for i in range(gap, n):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ø–∏—Å–∫–∞, –Ω–∞—á–∏–Ω–∞—è —Å gap
            temp = lst[i]  # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞
            j = i  # –ó–∞–ø–æ–º–∏–Ω–∞–µ–º —Ç–µ–∫—É—â–∏–π –∏–Ω–¥–µ–∫—Å
            # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏ –≤–Ω—É—Ç—Ä–∏ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤ —Å —à–∞–≥–æ–º gap
            while j >= gap and lst[j - gap] > temp:  # –ü–æ–∫–∞ —ç–ª–µ–º–µ–Ω—Ç—ã –≤–Ω—É—Ç—Ä–∏ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤ –±–æ–ª—å—à–µ —Ç–µ–∫—É—â–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
                lst[j] = lst[j - gap]  # –ü–µ—Ä–µ–º–µ—â–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –≤–Ω—É—Ç—Ä–∏ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤
                j -= gap  # –î–≤–∏–≥–∞–µ–º—Å—è –ø–æ –ø–æ–¥—Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º gap
            lst[j] = temp  # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –≤ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫
        k += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ k
        gap = 2 * (n // (2 ** (k + 1))) + 1  # –í—ã—á–∏—Å–ª—è–µ–º –Ω–æ–≤—ã–π —Ä–∞–∑–º–µ—Ä –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞
        if gap == 1:  # –ï—Å–ª–∏ —Ä–∞–∑–º–µ—Ä –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞ —Ä–∞–≤–µ–Ω 1
            if b == 0:  # –ï—Å–ª–∏ —Ñ–ª–∞–≥ –µ—â–µ –Ω–µ —Å–ª–æ–º–∞–Ω
                b += 1  # –õ–æ–º–∞–µ–º —Ñ–ª–∞–≥
            else:  
                break  # –ò–Ω–∞—á–µ –Ω–µ –ª–æ–º–∞–µ–º
    return lst  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫


# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
print("=" * 50)
primer_list = generate_random_list(10)
print(f'–ò–∑–Ω–∞—á–∞–ª—å–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Å–æ —Å–ª—É—á–∞–π–Ω—Ñ–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏: {primer_list}')
sorted_arr = shell_sort_modified(primer_list)  # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞
print(f'–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫: {sorted_arr}')  # –í—ã–≤–æ–¥ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞
```
