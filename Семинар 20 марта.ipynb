Класс – десятичная дробь. Методы класса – сумма, разность, произведение, частное десятичных дробей. Поля класса – целая и дробная часть числа.

```python
class DecimalFraction:
    def __init__(self, integer, fraction):
        self.integer = integer
        self.fraction = fraction

    def sum(self, other):
        sum_integer = self.integer + other.integer
        sum_fraction = self.fraction + other.fraction
        if sum_fraction >= 1:
            sum_integer += 1
            sum_fraction -= 1
        return DecimalFraction(sum_integer, sum_fraction)

    def difference(self, other):
        if self.fraction < other.fraction:
            difference_integer = self.integer - other.integer - 1
            difference_fraction = self.fraction + 1 - other.fraction
        else:
            difference_integer = self.integer - other.integer
            difference_fraction = self.fraction - other.fraction
        return DecimalFraction(difference_integer, difference_fraction)

    def product(self, other):
        product_integer = self.integer * other.integer
        product_fraction = self.fraction * other.fraction
        if product_fraction >= 1:
            product_integer += 1
            product_fraction -= 1
        return DecimalFraction(product_integer, product_fraction)

    def quotient(self, other):
        quotient_integer = self.integer / other.integer
        quotient_fraction = self.fraction / other.fraction
        if quotient_fraction >= 1:
            quotient_integer += 1
            quotient_fraction -= 1
        return DecimalFraction(quotient_integer, quotient_fraction)
```

Опишите класс Triangle, заданный длинами сторон. Включите в описание класса методы, позволяющие вывести длины сторон треугольника на экран, рассчитать периметр, площадь и высоты треугольника и свойство, позволяющее установить, существует ли треугольник с данными длинами сторон.

Класс Triangle:

- Свойства:
    - side1 - длина первой стороны треугольника
    - side2 - длина второй стороны треугольника
    - side3 - длина третьей стороны треугольника

- Методы:
    - print_sides() - выводит на экран длины сторон треугольника
    - calculate_perimeter() - рассчитывает и возвращает периметр треугольника
    - calculate_area() - рассчитывает и возвращает площадь треугольника
    - calculate_heights() - рассчитывает и возвращает высоты треугольника
    - is_triangle() - проверяет, существует ли треугольник с данными длинами сторон и возвращает булевое значение


```python
import math

class Triangle:
     # Конструктор класса для инициализации трех сторон треугольника
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    # Метод для вывода длин сторон треугольника
    def display_sides(self):  
        print("Длины сторон треугольника: ", self.a, self.b, self.c)
        
    # Метод для вычисления периметра треугольника
    def calculate_perimeter(self):
        perimeter = self.a + self.b + self.c
        return perimeter
    
    # Метод для вычисления площади треугольника по формуле Герона
    def calculate_area(self):
        s = (self.a + self.b + self.c) / 2
        area = math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
        return area
    
     # Метод для вычисления высот треугольника из каждой стороны
    def calculate_height(self):
        s = (self.a + self.b + self.c) / 2
        height_a = 2 * (s * (s - self.a) * (s - self.b) * (s - self.c)) / self.a
        height_b = 2 * (s * (s - self.a) * (s - self.b) * (s - self.c)) / self.b
        height_c = 2 * (s * (s - self.a) * (s - self.b) * (s - self.c)) / self.c
        return height_a, height_b, height_c
    
    # Метод для проверки существования треугольника по условию суммы двух сторон
    def check_existence(self):
        if (self.a + self.b > self.c) and (self.a + self.c > self.b) and (self.b + self.c > self.a):
            return True
        else:
            return False

# Создаем объект класса Triangle
triangle = Triangle(3, 4, 5)

# Выводим длины сторон треугольника
triangle.display_sides()

# Рассчитываем и выводим периметр треугольника
print("Периметр треугольника:", triangle.calculate_perimeter())

# Рассчитываем и выводим площадь треугольника
print("Площадь треугольника:", triangle.calculate_area())

# Рассчитываем и выводим высоты треугольника
print("Высоты треугольника:", triangle.calculate_height())

# Проверяем существует ли треугольник
print("Треугольник с данными сторонами существует:", triangle.check_existence())
```

Создайте класс ИЗДАНИЕ с методом, позволяющим вывести на экран информацию об издании, а также определить, является ли данное издание искомым. Создайте дочерние классы КНИГА (название, фамилия автора, год издания, издательство), СТАТЬЯ (название, фамилия автора, название журнала, его номер и год издания), ЭЛЕКТРОННЫЙ РЕСУРС (название, фамилия автора, ссылка, аннотация) со своими методами вывода информации на экран. Создайте список из  n  изданий, выведите полную информацию из списка, а также организуйте поиск изданий по фамилии автора.

```python
# Создаем класс "Издание"
class Publication:
    # Инициализируем атрибуты класса
    def __init__(self, title, author, year):
        self.title = title # Название издания
        self.author = author # Фамилия автора
        self.year = year # Год издания

    # Метод для вывода информации об издании на экран
    def show_info(self):
        print("Название издания:", self.title)
        print("Автор:", self.author)
        print("Год издания:", self.year)

    # Метод для определения, является ли данное издание искомым
    def is_searched(self, searched_author):
        # Проверяем, совпадает ли фамилия автора издания с искомой фамилией
        if self.author == searched_author:
            print("Данное издание является искомым.")
        else:
            print("Данное издание не является искомым.")

# Создаем дочерний класс "Книга"
class Book(Publication):
    # Добавляем новый атрибут - издательство
    def __init__(self, title, author, year, publisher):
        super().__init__(title, author, year)
        self.publisher = publisher # Издательство

    # Переопределяем метод вывода информации на экран для книги
    def show_info(self):
        print("Название книги:", self.title)
        print("Автор:", self.author)
        print("Издательство:", self.publisher)
        print("Год издания:", self.year)

# Создаем дочерний класс "Статья"
class Article(Publication):
    # Добавляем новые атрибуты - название журнала и его номер
    def __init__(self, title, author, year, journal, number):
        super().__init__(title, author, year)
        self.journal = journal # Название журнала
        self.number = number # Номер журнала

    # Переопределяем метод вывода информации на экран для статьи
    def show_info(self):
        print("Название статьи:", self.title)
        print("Автор:", self.author)
        print("Название журнала:", self.journal)
        print("Номер журнала:", self.number)
        print("Год издания:", self.year)

# Создаем дочерний класс "Электронный ресурс"
class ElectronicResource(Publication):
    # Добавляем новый атрибут - ссылка
    def __init__(self, title, author, year, link, annotation):
        super().__init__(title, author, year)
        self.link = link # Ссылка
        self.annotation = annotation # Аннотация

    # Переопределяем метод вывода информации на экран для электронного ресурса
    def show_info(self):
        print("Название ресурса:", self.title)
        print("Автор:", self.author)
        print("Ссылка:", self.link)
        print("Аннотация:", self.annotation)
        print("Год издания:", self.year)

# Создаем список из n изданий
publications = [Book("Война и мир", "Толстой", 1869, "Русский вестник"),
                Book("Преступление и наказание", "Достоевский", 1866, "Русский вестник"),
                Article("Искусство программирования", "Кнут", 1968, "Коммунар", 1),
                Article("Похождения бравого солдата Швейка", "Гаас", 1923, "Новый мир", 3),
                ElectronicResource("Python для начинающих", "Иванов", 2018, "httpsNO LINKSexampleNO LINKS", "Описание курса")]

# Выводим полную информацию из списка на экран
for publication in publications:
    print("Информация об издании:")
    publication.show_info()
    print()

# Организуем поиск изданий по фамилии автора
print("Поиск изданий по фамилии автора:")
searched_author = "Толстой"
for publication in publications:
    # Вызываем метод для определения, является ли издание искомым
    publication.is_searched(searched_author)
```
